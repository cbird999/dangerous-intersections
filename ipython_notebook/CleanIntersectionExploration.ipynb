```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn import datasets
```

```python
#READ IN DATA
intersections = pd.read_csv('../crash_data/clean_intersections.csv')
print intersections.columns
intersections.sort('name', ascending=False).head()
```

```python
# ADD NORMALIZED LAT LON
max_lat = intersections.lat.max()
min_lat = intersections.lat.min()
max_lon = intersections.lon.max()
min_lon = intersections.lon.min()
#i = 6
#print max_lat, min_lat, max_lon, min_lon
#print '---'
#print 'orig', intersections.lat.iget(i)
#print 'top', (intersections.lat.iget(i)- min_lat)
#print 'bottom', (max_lat - min_lat)
#print 'norm', (intersections.lat.iget(i)- min_lat)/(max_lat - min_lat)

intersections["lat_norm"] = (intersections.lat - min_lat) / (max_lat - min_lat)
intersections["lon_norm"] = (intersections.lon - min_lon) / (max_lon - min_lon)

#intersections.lon_norm.hist(bins=np.arange(0.0, 1.0, 0.1))
#intersections.lat.hist(bins=np.arange(40.0, 45.0, 0.1))
intersections.to_csv('../crash_data/clean_intersections.csv', index=False)
```

```python
#RUN SOME CLUSTER ALGORITHMS
import pickle 

learning_columns = filter(lambda x: x not in ['name', 'lat', 'lon', 'street1', 'street2', 'month',
                                              'year', 'borocode'],
                          intersections.columns)
for i in range(3, 9):
    k_means = KMeans(n_clusters=i).fit(intersections[learning_columns])
    pickle.dump(k_means, open('../mapping/clusters/{0}_means.pickle'.format(i), 'w'))
    print "dumping", i, "means"
    
intersections['colors'] = k_means.labels_
print 'inertia', k_means.inertia_
```

```python
#EXPLORE THE RESULT
#print 'cluster center', k_means.cluster_centers_
centers = k_means.cluster_centers_
n_clusters = len(centers)
for i in range(len(centers[0])):
    #print learning_columns[i], map(lambda x: round(x, 4), [centers[n,i] for n in range(n_clusters)])
    plt.figure()
    plt.bar(np.arange(-0.5, n_clusters-0.5) ,map(lambda x: round(x, 4), [centers[n,i] for n in range(n_clusters)]))
    plt.xlabel("cluster number")
    plt.ylabel("value")
    plt.title(learning_columns[i])
    plt.xticks(np.arange(n_clusters))

#print 'cluster labels', k_means.labels_
```

```python
#VISUALIZE CLUSTERS
colors = map(lambda x: ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'][x%8], k_means.labels_)
plt.scatter(intersections.lon, intersections.lat, c= colors, linewidth=0, alpha=0.1, s=1)
```
