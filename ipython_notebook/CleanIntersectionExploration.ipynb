```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn import datasets

from os import path
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#READ IN DATA
intersections = pd.read_csv('../crash_data/clean_intersections.csv')
print intersections.columns
intersections.sort('name', ascending=False).head()
```

```python
# ADD NORMALIZED LAT LON
max_lat = intersections.lat.max()
min_lat = intersections.lat.min()
max_lon = intersections.lon.max()
min_lon = intersections.lon.min()
intersections["lat_norm"] = (intersections.lat - min_lat) / (max_lat - min_lat)
intersections["lon_norm"] = (intersections.lon - min_lon) / (max_lon - min_lon)

intersections.to_csv('../crash_data/clean_intersections.csv', index=False)
```

```python
#RUN SOME CLUSTER ALGORITHMS
import pickle 

learning_columns = filter(lambda x: x not in ['name', 'lat', 'lon', 'street1', 'street2', 'month', 'year', 'borocode', 
                                               'lat_norm', 'lon_norm', 'colors', 'cluster', 'precinct',
                                               'err_confusn_ped_bike_other_ped', 'unsafe_speed', 'other_uninvolved_vehicle',
                                               'failure_to_keep_right', 'driver_inattention_distraction'], intersections.columns)
k_means = KMeans(n_clusters=8).fit(intersections[learning_columns])
    
intersections['cluster'] = k_means.labels_
print learning_columns
print 'done.'
```

```python
#VISUALIZE CLUSTERS
cols = ['m', 'k', 'g', 'y', 'b', 'r', 'c', 'w']
#random.shuffle(cols)
print cols
colors = map(lambda x: cols[x%8], k_means.labels_)
plt.figure(figsize=(15,15))
plt.scatter(intersections.lon, intersections.lat, c= colors, linewidth=0, alpha=0.5, s=5)
plt.xlim(min_lon, max_lon)
plt.ylim(min_lat, max_lat)
```

```python
#BUILD A DICT FOR COLORING
centers = k_means.cluster_centers_
n_clusters = len(centers)
clusters = range(n_clusters)

var_to_num = {}
var_to_cluster = {}
for c in range(len(centers[0])):  # go through var
    #find the centers for this particular variable
    center = [centers[cluster,c] for cluster in clusters]
    center.sort()
    top = center[-1]
    second = center[-2]
    perc_diff = ((top-second)/second)
    
    #save var for cluster
    var = learning_columns[c]
    #print center, max(center)
    center = [centers[cluster,c] for cluster in clusters]
    cluster = center.index(max(center))
    num = perc_diff
    #print var, perc_diff, cluster, "|", center
    var_to_num[var] = int(num*100)
    var_to_cluster[var] = cluster
        
#for var in learning_columns:
#    print var, var_to_cluster[var], var_to_num[var]
```

```python
#MAKE A WORD CLOUD TO DESCRIBE CLUSTERS
#text = "green green green green green green green purple"
text = ""
for var in learning_columns:
    #print var, var_to_cluster[var], var_to_num[var]
    text += (" "+var)*var_to_num[var]
cols = ['magenta', 'black', 'green', 'yellow', 'blue', 'red', 'cyan', 'white']
wc = WordCloud(font_path='/usr/share/fonts/truetype/dejavu/DejaVuSans-BoldOblique.ttf', 
               background_color='white',
               width=1000,
               height=500)
word_cloud = wc.generate(text)

def recolor(x1,x2,x3,x4,x5):
    #print x1, x2, x3, x4, x5
    #print x1[0], var_to_cluster[x1[0]], cols[var_to_cluster[x1[0]]]
    return cols[var_to_cluster[x1[0]]]
        

#print 'text2', text[:50]
#word count, font size, position and orientation
word_cloud = word_cloud.recolor(color_func = recolor)
# Open a plot of the generated image.
#plt.ioff()
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
#plt.savefig('../web/images/wordcloud.png')
#print wordcloud
word_cloud.to_file('../web/images/wordcloud.png')
```

```python
#EXPLORE THE RESULT
#print 'cluster center', k_means.cluster_centers_
centers = k_means.cluster_centers_
n_clusters = len(centers)
for i in range(len(centers[0])):
    #print learning_columns[i], map(lambda x: round(x, 4), [centers[n,i] for n in range(n_clusters)])
    plt.figure()
    plt.bar(np.arange(-0.5, n_clusters-0.5) ,[centers[n,i] for n in range(n_clusters)])
    plt.xlabel("cluster number")
    plt.ylabel("value")
    plt.title(learning_columns[i])
    plt.xticks(np.arange(n_clusters))

#print 'cluster labels', k_means.labels_
```
