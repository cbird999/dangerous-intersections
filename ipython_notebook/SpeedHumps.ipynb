```python
# ADD LAT / LON TO SPEED HUMPS
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import time

# Example: METROPOLITAN AVENUE and 74 AVENUE
def latlon_geocode(latlon):
    r = requests.get('http://nominatim.openstreetmap.org/search?q='+latlon+'&addressdetails=1&format=json')
    return r.json()

def address_geocode(address):
    r = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address='+address)
    return r.json()


def humps_at(a):
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address='+a+',new york, ny'
    r = requests.get(url)
    geo = r.json()
    time.sleep(0.5)
    
    lat = geo['results'][0]['geometry']['location']['lat']
    lon = geo['results'][0]['geometry']['location']['lng']
    return lat,lon

humps = pd.read_csv('../intervention_data/speed_humps.csv')
humps["STREETS"] = humps["FIRST_Main"]+"+"+humps["FIRST_ToSt"]
humps["LATLON"] = humps["STREETS"].apply(humps_at)
humps.to_csv("../intervention_data/speed_humps.csv", index=False)
humps.head()
```

```python
# ADD CRASH NUMBERS TO HUMPS
import pandas.tseries.offsets as offsets

def before_after(hump):
    # 1. find all crashes at this hump
    hump_street1 = hump["FIRST_Main"]
    hump_street2 = hump['FIRST_ToSt']
    c = crashes[ ((crashes['ON STREET NAME']==hump_street1) & (crashes['CROSS STREET NAME'] == hump_street2)) |    
                 ((crashes['CROSS STREET NAME']== hump_street1) & (crashes['ON STREET NAME'] == hump_street2)) ]
    
    
    # 2. select the crashes that occur at our two intervals
    s = {}
    date = pd.to_datetime(hump.FIRST_Inst, coerce=True)
    for interval in [3, 6]:
        before = date + offsets.DateOffset(days=30*interval)
        after = date
        s["CRASHES "+str(interval)] = 1.0*len( c[(pd.to_datetime(c.DATE, coerce=True) > after) &
                                             (pd.to_datetime(c.DATE, coerce=True) < before) ] )
    for interval in [-3, -6]:
        before = date 
        after = date - offsets.DateOffset(days=30*-interval)
        s["CRASHES "+str(interval)] = 1.0*len( c[(pd.to_datetime(c.DATE, coerce=True) > after) &
                                             (pd.to_datetime(c.DATE, coerce=True) < before) ] )
    return pd.Series(s)
    """ 
    # VERIFY
    print '----start---'
    c["DIFF"] = pd.to_datetime(c.DATE) - date
    print "crashes:", len(c)
    print "hump date:", date
    print c[["DATE", "DIFF"]]
    c.drop("DIFF", axis=1)
    print s
    print '----end---'
    
    """

# initially add the data columns and apply function
crashes = pd.read_csv("../crash_data/crashes.csv")
humps = pd.read_csv("../intervention_data/speed_humps.csv")
humps["CRASHES 3"] = -1
humps["CRASHES 6"] = -1
humps["CRASHES -3"] = -1
humps["CRASHES -6"] = -1
humps[["CRASHES 3", "CRASHES 6", "CRASHES -3", "CRASHES -6"]] = humps.apply(before_after, axis=1)

# add percent decrease columns
humps["ONES"] = 1.0

humps["3 MONTH PERCENT DECREASE"] = (humps["CRASHES -3"] - humps["CRASHES 3"]) / humps[["CRASHES -3", "ONES"]].max(axis=1)
humps["6 MONTH PERCENT DECREASE"] = (humps["CRASHES -6"] - humps["CRASHES 6"]) / humps[["CRASHES -6", "ONES"]].max(axis=1)
humps.drop(["ONES"], axis=1, inplace=True)
humps.to_csv("../intervention_data/speed_humps.csv", index=False)
humps.sort("CRASHES 3", ascending=False).head(n=10)
```
