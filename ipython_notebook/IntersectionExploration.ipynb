INTERSECTIONS: RELATIONSHIP BETWEEN CRASHES AND WALKABILITY? 
---------------------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#What are the top [n] intersections for [crashes]?
intersections = pd.read_csv('../crash_data/intersections.csv')
intersections.head()
#intersections.sort('CRASHES', ascending=False, inplace=True)
#intersections[["LATITUDE", "LONGITUDE", "ON STREET NAME", "OFF STREET NAME", "CROSS STREET NAME", "CRASHES"]].head(n=10)
```

```python
intersections.CRASHES.hist(bins=np.arange(0,100,5))
#intersections.CRASHES.apply(np.log).hist(bins=np.arange(0,100,5))
```

```python
# What is the distribution of number of businesses?
intersections.YELP_FOOD.hist(bins=np.arange(-0.5,20))
```

```python
# How well can number of crashes predict amount of pedestrian traffic?
intersections["HIGH_PD_TRAFFIC"] = (1.0*(intersections.YELP_FOOD > 0))
plt.scatter(intersections.CRASHES, intersections.HIGH_PD_TRAFFIC)
plt.xlabel("Number of crashes")
plt.ylabel("High Pedestrian traffic (boolean)")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```

```python
# How well number of businesses predict traffic crashes?
model = sm.OLS(intersections.CRASHES, intersections.YELP_FOOD).fit()
print 'model summary', model.summary()

# plot it
plt.scatter(intersections.YELP_FOOD, intersections.CRASHES, linewidth=0, alpha=0.2)
plt.plot(np.arange(0,20), model.predict(np.arange(0,20)), 'black')
plt.ylabel("Number of crashes")
plt.xlabel("Number Yelp businesses")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```

```python
# How well can number of crashes predict amount of pedestrian deaths / injuries?
#y = intersections["NUMBER OF PEDESTRIANS INJURED"]
y = intersections["NUMBER OF PEDESTRIANS KILLED"]
x = intersections.YELP_FOOD
model = sm.OLS(y, x).fit()
print 'model summary', model.summary()

# plot it
plt.scatter(x, y, linewidth=0, alpha=0.2)
plt.plot(np.arange(0,20), model.predict(np.arange(0,20)), 'black')
plt.ylabel("Number of pedestrians killed")
plt.xlabel("Number Yelp businesses")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```

```python
#WHAT DO FOUR SQUARE CHECK INS LOOK LIKE
intersections["FOUR_SQUARE_FOOD"].apply(lambda x: 20*np.log(x)).hist(bins=np.arange(0, 300, 5))
```

```python
# How well can four square predict traffic crashes?
predictor = intersections["FOUR_SQUARE_FOOD"].apply(lambda x: np.log(max((1,x))))
#print predictor
model = sm.OLS(intersections.CRASHES, predictor).fit()
print 'model summary', model.summary()
#print 'model predictions', model.predict(np.arange(0,20))
# plot it
plt.scatter(predictor, intersections.CRASHES, linewidth=0, alpha=0.2)
plt.plot(np.arange(0,20), model.predict(np.arange(0,20)), 'black')
plt.ylabel("Number of crashes")
plt.xlabel("Four Square checkins")
plt.title('How well can four square checkins predict\n amount of pedestrian traffic?')
```

```python
# Visualizing a color block
X = []
Y = []
cols = []
for x in np.arange(0,255,1):
    for y in np.arange(0,255,1):
        X.append(x)
        Y.append(y)
        cols.append('#%02x%02x%02x' % (x, y, 0))
plt.figure(figsize=(10,10))
plt.scatter(X,Y,c=cols, linewidth=0)
plt.xlim(0,255)
plt.ylim(0,255)
plt.ylabel('number of crashes',fontsize=40)
plt.xlabel('number of businesses',fontsize=40)
```
