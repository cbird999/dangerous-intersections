```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
crashes = pd.read_csv('crash_data/crashes.csv', names=['DATE', 'TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE', 'LONGITUDE', 
                                                       'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
                                                       'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
                                                       'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
                                                       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
                                                       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
                                                       'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2',
                                                       'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 
                                                       'CONTRIBUTING FACTOR VEHICLE 5', 'UNIQUE KEY', 'VEHICLE TYPE CODE 1',
                                                       'VEHICLE TYPE CODE 2', 'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4',
                                                       'VEHICLE TYPE CODE 5'])
crashes.head(n=25)
```

```python
# HOW MANY ARE INJURED / KILLED PER CRASH
total_crashes = float(len(crashes))
print "FIRST CRASH", np.min(crashes["DATE"])
print "TOTAL CRASHES", total_crashes
print '\t\t\tNUMBER\tRATE'
for death in ['PERSONS INJURED', 'PERSONS KILLED',
               'PEDESTRIANS INJURED', 'PEDESTRIANS KILLED',
               'CYCLIST INJURED', 'CYCLIST KILLED',
               'MOTORIST INJURED', 'MOTORIST KILLED']:
    total = np.sum(crashes['NUMBER OF ' + death])
    print death,'  \t', total, "\t", total/total_crashes
```

# SPEED HUMPS: HOW EFFECTIVE ARE SPEED HUMPS?

```python
# HISTOGRAMS OF PERCENT DECREASE BEFORE AND AFTER HUMP INSTALLATION
humps = pd.read_csv('../intervention_data/speed_humps.csv')
city_pd = humps[['CRASHES 3','CRASHES -3','CRASHES 6','CRASHES -6']].sum(axis=0)
print city_pd
print '3 month city wide pd', (city_pd["CRASHES -3"] - city_pd["CRASHES 3"]) / city_pd[["CRASHES -3", "ONES"]].max(axis=1)
print '6 month city wide pd', (city_pd["CRASHES -6"] - city_pd["CRASHES 6"]) / city_pd[["CRASHES -6", "ONES"]].max(axis=1)
print '-----------'
for i in [3, 6]:
    plt.figure()
    plt.hist(humps[str(i) + ' MONTH PERCENT DECREASE'], np.arange(-3, 3, 0.2))
    m = humps[str(i) + ' MONTH PERCENT DECREASE'].mean()
    sd = humps[str(i) + ' MONTH PERCENT DECREASE'].sem()
    print i, "mean", round(m, 3), "| 95% CI", (m-(2*sd), m+(2*sd))
    plt.plot([m, m], [0, 120], 'r')
    
    plt.xlabel('Percent decrease')
    plt.ylabel('Frequency')
    plt.title('Percent decrease in crashes over ' + str(i) + ' months.')
```

```python
# DOES CB AFFECT SPEED HUMP EFFECTIVENESS
humps["CB_BO"] = humps.FIRST_CB.apply(str)+" " + humps.FIRST_Boro.apply(str)
cb_gb = humps.groupby('CB_BO', as_index=False)
cb_agg = cb_gb.aggregate(np.sum)
cb_agg["ONES"] = 1.0
cb_agg["3 MONTH PERCENT DECREASE"] = (cb_agg["CRASHES -3"] - cb_agg["CRASHES 3"]) / cb_agg[["CRASHES -3", "ONES"]].max(axis=1)
cb_agg["6 MONTH PERCENT DECREASE"] = (cb_agg["CRASHES -6"] - cb_agg["CRASHES 6"]) / cb_agg[["CRASHES -6", "ONES"]].max(axis=1)
cb_agg.drop(["ONES"], axis=1, inplace=True)
cb_agg[["CB_BO","CRASHES -3", "CRASHES 3", 'CRASHES -6', 'CRASHES 6', "3 MONTH PERCENT DECREASE", "6 MONTH PERCENT DECREASE"]].sort("3 MONTH PERCENT DECREASE", ascending=False).head(n=10)
```

```python
# DOES BORO AFFECT SPEED HUMP EFFECTIVENESS
"""
Borough numbers: 
    1. Manhattan (New York County)
    2. Bronx (Bronx County)
    3. Brooklyn (Kings County)
    4. Queens (Queens County)
    5. Staten Island (Richmond County)
"""
bo_gb = humps.groupby('FIRST_Boro', as_index=False)
bo_agg = bo_gb.aggregate(np.sum)
bo_agg["ONES"] = 1.0
bo_agg["3 MONTH PERCENT DECREASE"] = (bo_agg["CRASHES -3"] - bo_agg["CRASHES 3"]) / bo_agg[["CRASHES -3", "ONES"]].max(axis=1)
bo_agg["6 MONTH PERCENT DECREASE"] = (bo_agg["CRASHES -6"] - bo_agg["CRASHES 6"]) / bo_agg[["CRASHES -6", "ONES"]].max(axis=1)
bo_agg.drop(["ONES"], axis=1, inplace=True)
bo_agg[["FIRST_Boro","CRASHES -3", "CRASHES 3", 'CRASHES -6', 'CRASHES 6', "3 MONTH PERCENT DECREASE", "6 MONTH PERCENT DECREASE"]]\
    .sort("3 MONTH PERCENT DECREASE", ascending=False)
```

# INTERSECTIONS:  RELATIONSHIP BETWEEN CRASHES AND WALKABILITY?

```python
intersections = pd.read_csv('../crash_data/intersections.csv')
intersections.head()
```

```python
# What is the distribution of number of businesses?
intersections.YELP80.hist(bins=np.arange(-0.5,20))
```

```python
# How well can number of crashes predict amount of pedestrian traffic?
intersections["HIGH_PD_TRAFFIC"] = (1.0*(intersections.YELP80 > 0))
plt.scatter(intersections.CRASHES, intersections.HIGH_PD_TRAFFIC)
plt.xlabel("Number of crashes")
plt.ylabel("High Pedestrian traffic (boolean)")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```

```python
# How well number of businesses predict traffic crashes?
model = sm.OLS(intersections.CRASHES, intersections.YELP80).fit()
print 'model summary', model.summary()

# plot it
plt.scatter(intersections.YELP80, intersections.CRASHES, linewidth=0, alpha=0.2)
plt.plot(np.arange(0,20), model.predict(np.arange(0,20)), 'black')
plt.ylabel("Number of crashes")
plt.xlabel("Number Yelp businesses")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```

```python
# How well can number of crashes predict amount of pedestrian deaths / injuries?
#y = intersections["NUMBER OF PEDESTRIANS INJURED"]
y = intersections["NUMBER OF PEDESTRIANS KILLED"]
x = intersections.YELP80
model = sm.OLS(y, x).fit()
print 'model summary', model.summary()

# plot it
plt.scatter(x, y, linewidth=0, alpha=0.2)
plt.plot(np.arange(0,20), model.predict(np.arange(0,20)), 'black')
plt.ylabel("Number of crashes")
plt.xlabel("Number Yelp businesses")
plt.title('How well can number of crashes predict\n amount of pedestrian traffic?')
```
